# Energy Trends in the US from 1973 - 2013

###### Data Set Source:http://www.eia.gov/

##### Abstract: At the turn of the Century, the US was decisively a major consumer in the energy sector and yet, 14 years later it is poised to become a Global  energy producer thanks to a renewed interest in domestic fosil fuel supplies as well as a growing renewable and or carbon neutral energy sector.  But just how has this influx of "new" energy affected the energy sector and does this current upswing fall within the norms of past consumption cycles, or does it break with previous trends?   Analyzing data going as far back as 1973, we'll be able to look at the growth cycles of the energy sector for a multitude of energy sources ranging from coal, oil, natual gas, nuclear power, renewable sources, etc. In addition to growth cycles, we will also be able to see the consumption rates and cylces for these various resources going just as far back as well. There is also one other aspect of the energy sector that will be explored and that aspect is electrical generation and consumption.


This Project uses Data Sets for Renewable Energy, Nuclear Energy and Fossil Fuels as well as a Primary Data Set for overall Energy Use in the US.

Motivation:  To see how the different energy sectors in the US affect each other and the overall use of energy in the states
